```python
import os
import numpy as np   
import pandas as pd
import calendar
import locale
import datetime
```

```python
# Leer los archivos csv
df_2010 = pd.read_csv("./IMIDA_dia_2010.csv", encoding='latin1', sep=';')
df_2011 = pd.read_csv("./IMIDA_dia_2011.csv", encoding='latin1', sep=',')
df_2012 = pd.read_csv("./IMIDA_dia_2012.csv", encoding='latin1', sep=',')
df_2013 = pd.read_csv("./IMIDA_dia_2013.csv", encoding='latin1', sep=',')
df_2014 = pd.read_csv("./IMIDA_dia_2014.csv", encoding='latin1', sep=',')
df_2015 = pd.read_csv("./IMIDA_dia_2015.csv", encoding='latin1', sep=',')
df_2016 = pd.read_csv("./IMIDA_dia_2016.csv", encoding='latin1', sep=',')
df_2017 = pd.read_csv("./IMIDA_dia_2017.csv", encoding='latin1', sep=',')
df_2018 = pd.read_csv("./IMIDA_dia_2018.csv", encoding='latin1', sep=';')
df_2019 = pd.read_csv("./IMIDA_dia_2019.csv", encoding='latin1', sep=';')
df_2020 = pd.read_csv("./IMIDA_dia_2020.csv", encoding='latin1', sep=';')
```

```python
# Establece el idioma a español
locale.setlocale(locale.LC_TIME, 'es_ES.UTF-8')

# Convertir la columna "FECHA" en formato de fecha
df_2010["FECHA"] = pd.to_datetime(df_2010["FECHA"], format="%d/%m/%Y", dayfirst=True)
df_2011["FECHA"] = pd.to_datetime(df_2011["FECHA"], format="%d/%m/%y", dayfirst=True)
df_2012["FECHA"] = pd.to_datetime(df_2012["FECHA"], format="%d/%m/%y", dayfirst=True)
df_2013["FECHA"] = pd.to_datetime(df_2013["FECHA"], format="%d/%m/%y", dayfirst=True)
df_2014["FECHA"] = pd.to_datetime(df_2014["FECHA"], format="%d/%m/%y", dayfirst=True)
df_2015["FECHA"] = pd.to_datetime(df_2015["FECHA"], format="%d/%m/%y", dayfirst=True)
df_2016["FECHA"] = pd.to_datetime(df_2016["FECHA"], format="%d/%m/%y", dayfirst=True)
df_2017["FECHA"] = pd.to_datetime(df_2017["FECHA"], format="%d/%m/%y", dayfirst=True)
df_2018["FECHA"] = pd.to_datetime(df_2018["FECHA"], format="%d/%m/%Y", dayfirst=True, errors='coerce')
df_2019["FECHA"] = pd.to_datetime(df_2019["FECHA"], format="%d/%m/%Y", dayfirst=True)
df_2020["FECHA"] = pd.to_datetime(df_2020["FECHA"], format="%d/%m/%Y", dayfirst=True)
```

```python
# Crear la nueva columna "MES" con el nombre del mes correspondiente en español
df_2010["MES"] = df_2010["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
df_2011["MES"] = df_2011["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
df_2012["MES"] = df_2012["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
df_2013["MES"] = df_2013["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
df_2014["MES"] = df_2014["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
df_2015["MES"] = df_2015["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
df_2016["MES"] = df_2016["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
df_2017["MES"] = df_2017["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
df_2018["MES"] = df_2018["FECHA"].dt.month.fillna(0).apply(lambda x: calendar.month_name[int(x)]).str.capitalize()
df_2019["MES"] = df_2019["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
df_2020["MES"] = df_2020["FECHA"].dt.month.apply(lambda x: calendar.month_name[x]).str.capitalize()
```

```python
# Crear la nueva columna "DIA_DEL_AÑO"
df_2012["DIA_DEL_AÑO"] = df_2012["FECHA"].dt.dayofyear
```

```python
# Lista con todos los DataFrames
dfs = [df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020]

# Iterar sobre cada DataFrame en la lista
for df in dfs:
    
    # Crear la nueva columna "DIA_DEL_AÑO"
    df["DIA_DEL_AÑO"] = df["FECHA"].dt.dayofyear
```

```python
# Juntar todos los DataFrames en uno solo
dfs = [df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020]
df_total = pd.concat(dfs)
```

```python
df_total.head(180000)
```

```python
# Convertir las columnas a numéricas
columnas = ['TMED', 'TMAX', 'TMIN', 'HRMED', 'HRMAX', 'HRMIN', 'PREC', 'RADMED', 'VVMED', 'VVMAX', 'DVMED', 'ETO']

for columna in columnas:
    df_total[columna] = pd.to_numeric(df_total[columna], errors='coerce')
```

```python
# Elimina las columnas 'FECHA' y 'EST'
df_total = df_total.drop(['FECHA', 'EST'], axis=1)
```

```python
# Añade la columna 'RIEGO'
df_total['RIEGO'] = (
    ((df_total['TMED'] > 20) | (df_total['TMAX'] > 28)) &
    ((df_total['HRMED'] < 60) | (df_total['HRMIN'] < 40)) &
    (df_total['PREC'] < 1.0) &
    (df_total['RADMED'] > 200)
).astype(int)
```

```python
# Reemplazar errores en los nombres de la columna 'PARAJE'
df_total['PARAJE'] = df_total['PARAJE'].replace('Casas del Rey', 'Casas Del Rey')
df_total['PARAJE'] = df_total['PARAJE'].replace('Caï¿½ada De San Pedro', 'Cañada de San Pedro')
df_total['PARAJE'] = df_total['PARAJE'].replace('Caï¿½ada Gallego', 'Cañada Gallego')
df_total['PARAJE'] = df_total['PARAJE'].replace('Cehegï¿½n', 'Cehegín')
df_total['PARAJE'] = df_total['PARAJE'].replace('Cañada del Judio', 'Cañada Del Judío')
df_total['PARAJE'] = df_total['PARAJE'].replace('Cª Del Judio', 'Cañada Del Judío')
df_total['PARAJE'] = df_total['PARAJE'].replace('Cï¿½ Del Judio', 'Cañada Del Judío')
df_total['PARAJE'] = df_total['PARAJE'].replace('Llano de las olivericas', 'Llano De Las Olivericas')
df_total['PARAJE'] = df_total['PARAJE'].replace('Loma del Aire', 'Loma Del Aire')
df_total['PARAJE'] = df_total['PARAJE'].replace('Martï¿½nez Del Puerto', 'Martínez Del Puerto')
df_total['PARAJE'] = df_total['PARAJE'].replace('Santiago de la Ribera', 'Santiago De La Ribera')
df_total['PARAJE'] = df_total['PARAJE'].replace('Estaciï¿½n De Blanca', 'Estación De Blanca')
```

```python
# Marcar las columnas numéricas
numeric_columns = df_total.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Calcular la media aritmética de las columnas numéricas para cada grupo
df_pronostico = df_total.groupby(["PARAJE", "DIA_DEL_AÑO", "MES"], as_index=False)[numeric_columns].mean()
```

```python
df_pronostico.head(10000)
```

```python
# Redondear todos los números a dos decimales
df_pronostico = df_pronostico.round(2)
```

```python
# Ajustar formato de los campos numéricos
df_pronostico['DIA_DEL_AÑO'] = df_pronostico['DIA_DEL_AÑO'].astype(int)
# df_pronostico = df_pronostico.applymap(lambda x: format(x, '.2f').replace('.', ',') if isinstance(x, float) else x)
```

```python
# Diccionario para mapear los nombres de los meses a sus números correspondientes
meses = {'Enero': 1, 'Febrero': 2, 'Marzo': 3, 'Abril': 4, 'Mayo': 5, 'Junio': 6,
         'Julio': 7, 'Agosto': 8, 'Septiembre': 9, 'Octubre': 10, 'Noviembre': 11, 'Diciembre': 12}

# Crea la nueva columna 'NUMERO_MES'
df_pronostico['NUMERO_MES'] = df_pronostico['MES'].map(meses)

df_pronostico.head()
```

```python
def crear_dia_mes(row):
    mes = row['MES']
    dia_del_ano = row['DIA_DEL_AÑO']
    
    # Crear un objeto datetime para el primer día del año 2025
    primer_dia = datetime.datetime(2025, 1, 1)
    
    # Añadir el número de días al primer día del año
    fecha = primer_dia + datetime.timedelta(days=dia_del_ano - 1)
    
    # Formatear la fecha en el formato "dd-mm-yyyy"
    dia_mes = fecha.strftime('%d/%m/%Y')
    
    return dia_mes

# Aplicar la función a cada fila del DataFrame para crear la nueva columna "FECHA"
df_pronostico['FECHA'] = df_pronostico.apply(crear_dia_mes, axis=1)

# Mostrar el DataFrame con la nueva columna
df_pronostico.head()
```

```python
# Elimina las columnas 'DIA_DEL_AÑO' y 'NUMERO MES'
df_pronostico = df_pronostico.drop(['DIA_DEL_AÑO', 'NUMERO_MES'], axis=1)
```

```python
# Crear un diccionario para eliminar posibles valores duplicados
meses = {
    'Enero': '01/02/2025',
    'Febrero': '01/03/2025',
    'Marzo': '01/04/2025',
    'Abril': '01/05/2025',
    'Mayo': '01/06/2025',
    'Junio': '01/07/2025',
    'Julio': '01/08/2025',
    'Agosto': '01/09/2025',
    'Septiembre': '01/10/2025',
    'Octubre': '01/11/2025',
    'Noviembre': '01/12/2025',
    'Diciembre': '01/01/2026'
}

# Usar un bucle para filtrar el DataFrame para cada par de mes y fecha en el diccionario
for mes, fecha in meses.items():
    df_pronostico = df_pronostico[~((df_pronostico['MES'] == mes) & (df_pronostico['FECHA'] == fecha))]
```

```python
# Lista de columnas para convertir
columnas = ['TMED', 'TMAX', 'TMIN', 'HRMED', 'HRMAX', 'HRMIN', 'PREC', 'RADMED', 'VVMED', 'VVMAX', 'DVMED', 'ETO']

# Convertir cada columna a numérico
for columna in columnas:
    df_pronostico[columna] = pd.to_numeric(df_pronostico[columna], errors='coerce')

# Verificar el resultado
df_pronostico.dtypes
```
